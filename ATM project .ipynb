```python
# mini project

class MyAtm:
    def __init__(self):
        self.pin = 0
        self.balance = 0
        self.menu()

    def menu(self):
        choice = int(input("""
        Please choose from below options
        1. Create PIN
        2. Deposit
        3. Withdraw
        4. Check Balance
        5. Exit
        """))

        if choice == 1:
            self.create_pin()
            self.menu()
        elif choice == 2:
            self.deposit()
            self.menu()
        elif choice == 3:
            self.withdraw()
            self.menu()
        elif choice == 4:
            self.check_balance()
            self.menu()
        elif choice == 5:
            print("Thank you for banking with us!!!")
        else:
            print("Invalid option")

    def create_pin(self):
        self.pin = int(input("Create your PIN:"))
        self.balance = int(input("Enter your current balance:"))

    def deposit(self):
        amount = int(input("Enter amount to be deposited:"))
        user_pin = int(input("Enter your PIN:"))

        if user_pin == self.pin:
            self.balance = self.balance + amount
            print("Paise aa gaye")
        else:
            print("Incorrect PIN")

    def withdraw(self):
        amount = int(input("Enter amount to be withdrawn:"))
        user_pin = int(input("Enter your PIN:"))

        if user_pin == self.pin:
            if amount <= self.balance:
                self.balance = self.balance - amount
                print("send ")
            else:
                print("Garib")
        else:
            print("Incorrect PIN")

    def check_balance(self):
        user_pin = int(input("Enter your PIN:"))

        if user_pin == self.pin:
            print("Available balance", self.balance)
        else:
            print("Incorrect PIN")

#A=MyAtm()
```
